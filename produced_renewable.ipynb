# Clean the Data

```python
# import dependencies

import pandas as pd
```

```python
# read the data 
df = pd.read_csv("prod-data.csv")
pd.set_option('display.max_columns', None)
df
```

```python
# see the df details
df.info()
```

```python
# convert the data change columns to rows so our data is lenear

clean_df = df.melt(id_vars=["State"], 
        var_name="Year", 
        value_name="Data")
pd.set_option('display.max_rows', None)
clean_df
```

```python
# save the dataframe as csv
clean_df.to_csv("produced")
```
